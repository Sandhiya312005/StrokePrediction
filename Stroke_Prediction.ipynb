```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.25, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred_lr = lr.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
accuracy_lr = accuracy_score(ytest,ypred_lr)
accuracy_lr
```

```python
cl_rep_lr = classification_report(ytest , ypred_lr)
print(cl_rep_lr)
```

```python
con_mat_lr = confusion_matrix(ytest , ypred_lr)
print(con_mat_lr)
```

```python
import seaborn as sns
sns.countplot(data=data, x='stroke')
```

```python
from sklearn.utils import resample
#create two different dataframe of majority and minority class
df_majority = data[(data['stroke']==0)]
df_minority = data[(data['stroke']==1)]
# upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True,         # sample with replacement
                                 n_samples= 4861,      # to match majority class
                                 random_state=42)      # reproducible results
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
```

```python
import seaborn as sns
sns.countplot(data=df_upsampled, x='stroke')
```

```python
df_upsampled['stroke'].value_counts()
```

```python
df_upsampled.head()
```

```python
resample_x = df_upsampled.iloc[:,:-1]
resample_y = df_upsampled.iloc[:,-1:]
```

```python
resample_xtrain , resample_xtest , resample_ytrain , resample_ytest = train_test_split(resample_x , resample_y , test_size = 0.2 , random_state = 1)
```

```python
resample_lr = LogisticRegression()
resample_lr.fit(resample_xtrain , resample_ytrain)
```

```python
resample_ypred = resample_lr.predict(resample_xtest)
```

```python
resample_accuracy = accuracy_score(resample_ytest , resample_ypred)
resample_accuracy
```

```python
resample_confusion_matrix = confusion_matrix(resample_ytest , resample_ypred)
print(resample_confusion_matrix)
```

```python
resample_classification_report = classification_report(resample_ytest , resample_ypred)
print(resample_classification_report)
```

```python
from imblearn.over_sampling import SMOTE
# Resampling the minority class. The strategy can be changed as required.
sm = SMOTE(sampling_strategy='minority', random_state=42)
# Fit the model to generate the data.
oversampled_X, oversampled_Y = sm.fit_resample(data.drop('stroke', axis=1), data['stroke'])
oversampled = pd.concat([pd.DataFrame(oversampled_X), pd.DataFrame(oversampled_Y)], axis=1)
```

```python
oversampled.head()
```

```python
oversampled['stroke'].value_counts()
```
